```python
import json

with open(r'profile2.json', 'r') as f:
    profile_setting = json.load(f)

data_logpath = profile_setting["logPath"]
data_regex2 = profile_setting["parseRule"]["regex"]

column_list = []

for item in profile_setting["parseRule"]["item"]:
    column_list.append(item)
    
print(json.dumps(profile_setting, indent=4))
```

```python
import re
import pandas as pd

def parse_log(data_logpath, data_regex2, column_list, log_df):
    val_list=[]
    val_list_index=0
    row_num=0
    item_num = len(column_list)
    text = ""
    
    f = open(data_logpath)
    
    for line in f:
        if line != "\n":
            text += line
        else:
            data_regex = re.compile(data_regex2, re.DOTALL)
            log_regex = data_regex.search(text)
            if log_regex != None:
                print(log_regex.group(1) + " " + log_regex.group(2))
                if log_regex.group(1) == column_list[val_list_index]:
                    val_list.append(log_regex.group(2))
                    val_list_index += 1
                    if val_list_index == item_num:
                        log_df.loc[row_num] = val_list
                        val_list.clear()
                        row_num += 1
                        val_list_index = 0
                else:
                    val_list.append("8.8")
                    val_list_index += 1
                    if val_list_index == item_num:
                        log_df.loc[row_num] = val_list
                        val_list.clear()
                        val_list.append(log_regex.group(2))
                        row_num += 1
                        val_list_index = 1
                    else:
                        val_list.append(log_regex.group(2))
                        val_list_index += 1
                        if val_list_index == item_num:
                            log_df.loc[row_num] = val_list
                            val_list.clear()
                            row_num += 1
                            val_list_index = 0
                text = ""
    f.close()
    
    return log_df

log_df = pd.DataFrame(columns = column_list)
parse_log(data_logpath, data_regex2, column_list, log_df)
```
